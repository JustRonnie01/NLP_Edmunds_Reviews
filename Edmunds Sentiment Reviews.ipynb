```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Edmunds car review.csv')
```

```python
df.isnull().sum()
```

```python
df = df.drop('Reviewer', axis=1)
```

```python
df.head(25)
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df.isna().sum()
```

```python

df["Title"] = df["Title"].astype(str)
# Apply the custom function to create a new column 'Combined'
df['Combined_Title_Review'] = df[['Title', 'Review']].agg(' '.join, axis=1)
```

```python
df.head(25)
```

```python
df.info()
```

```python
# Create a new column 'Sentiment' based on conditions
df['Sentiment'] = 'Unknown'  # Default value

# Apply conditions for 'Positive'
df.loc[df['Rating'] >= 4, 'Sentiment'] = 'Positive'

# Apply conditions for 'Neutral'
df.loc[df['Rating'] == 3, 'Sentiment'] = 'Neutral'

# Apply conditions for 'Negative'
df.loc[df['Rating'] <= 2, 'Sentiment'] = 'Negative'
```

```python
df.head()
```

```python
df.head()
```

```python
df['Sentiment'].value_counts()
```

```python
y = df['Sentiment']

X = df['Combined_Title_Review']
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=23)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
tfidf.fit(X_train)
```

```python
X_train_tfidf = tfidf.transform(X_train)
```

```python
X_test_tfidf = tfidf.transform(X_test)
```

```python
X_train_tfidf
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
nb = MultinomialNB()
nb.fit(X_train_tfidf, y_train)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
```

```python
preds = nb.predict(X_test_tfidf)

print(classification_report(y_test,preds))

ConfusionMatrixDisplay(nb,X_test_tfidf, y_test)
```

```python

# Generate confusion matrix
cm = confusion_matrix(y_test, preds, labels=nb.classes_)

# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=nb.classes_)
disp.plot()

# Display classification report
print(classification_report(y_test, preds))
```
