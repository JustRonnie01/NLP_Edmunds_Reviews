```python
df["Title"] = df["Title"].astype(str)
# Apply the custom function to create a new column 'Combined'
df['Combined_Title_Review'] = df[['Title', 'Review']].agg(' '.join, axis=1)
```

```python
# Create a new column 'Sentiment' based on conditions
df['Sentiment'] = 'Unknown'  # Default value

# Apply conditions for 'Positive'
df.loc[df['Rating'] >= 4, 'Sentiment'] = 'Positive'

# Apply conditions for 'Neutral'
df.loc[df['Rating'] == 3, 'Sentiment'] = 'Neutral'

# Apply conditions for 'Negative'
df.loc[df['Rating'] <= 2, 'Sentiment'] = 'Negative'
```

```python
# Group by 'Company' and calculate the ratio of positive to negative reviews
reviews_by_brand = df.groupby(['Company', 'Sentiment']).size().unstack(fill_value=0)
reviews_by_brand['PositiveToNegativeRatio'] = reviews_by_brand['Positive'] / reviews_by_brand['Negative']

# Sort the data by the ratio in descending order
reviews_by_brand = reviews_by_brand.sort_values(by='PositiveToNegativeRatio', ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='PositiveToNegativeRatio', y=reviews_by_brand.index, data=reviews_by_brand, palette='viridis')

plt.title('Positive to Negative Reviews Ratio by Brand')
plt.xlabel('Positive to Negative Reviews Ratio')
plt.ylabel('Car Brand')

plt.show()
```

```python

# Group by 'Company' and count the number of negative reviews
negative_reviews_by_brand = df[df['Sentiment'] == 'Negative'].groupby('Company').size().sort_values(ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x=negative_reviews_by_brand.values, y=negative_reviews_by_brand.index, palette='Reds_r')

plt.title('Total Negative Reviews by Brand')
plt.xlabel('Number of Negative Reviews')
plt.ylabel('Car Brand')

plt.show()
```

```python


# Calculate the percentage of negative and positive reviews for each brand
total_reviews_by_brand = df['Company'].value_counts()
negative_reviews_by_brand = df[df['Sentiment'] == 'Negative']['Company'].value_counts()
positive_reviews_by_brand = df[df['Sentiment'] == 'Positive']['Company'].value_counts()

negative_percentage = (negative_reviews_by_brand / total_reviews_by_brand) * 100
positive_percentage = (positive_reviews_by_brand / total_reviews_by_brand) * 100

# Calculate the ratio of total negative to positive percentage reviews
ratio_percentage = negative_percentage / positive_percentage

# Plot the ratio of total negative to positive percentage reviews
plt.figure(figsize=(10, 6))
sns.barplot(x=ratio_percentage.values, y=ratio_percentage.index, palette='coolwarm')
plt.title('Ratio of Total Negative to Positive Percentage Reviews by Brand')
plt.xlabel('Ratio (Negative Percentage / Positive Percentage)')
plt.ylabel('Brand')
plt.show()

```

```python
# Sort the ratio_percentage Series in descending order
sorted_ratio_percentage = ratio_percentage.sort_values(ascending=True)

# Plot the sorted ratio of total negative to positive percentage reviews
plt.figure(figsize=(10, 6))
sns.barplot(x=sorted_ratio_percentage.values, y=sorted_ratio_percentage.index, palette='crest')
plt.title('Top Brands')
plt.xlabel('Ratio (Negative Percentage / Positive Percentage)')
plt.ylabel('Brand')
plt.show()
```
